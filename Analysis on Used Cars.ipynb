# **8. Exploratory Data Analysis**
---

- This section is emphasised on asking the right questions and perform analysis using the data.

- Note that there is no limit how deep you can go, but make sure not to get distracted from right track.

First lets see how many cars are sold per year.

```python
count_by_year = data.groupby("year").count()
count_by_year
```

So we see that we have data of cars from 1953 to 2016. Lets see the trend of number of cars being sold by year

```python
count_by_year[['car']].plot(kind='line', figsize=(15,7), color='blue', fontsize=13, linestyle='-.')

plt.xlabel('Year')
plt.ylabel('Count of Cars sold')
plt.title('Line chart showing the trend of number of cars sold per year')
```

Lets now see the preferences of buyers by year.

```python
count_by_year_and_body = data.groupby(["year","body"]).count()
count_by_year_and_body
```

Since the number of rows are 232, lets see the trend in a decade.

```python
fig,ax = plt.subplots(figsize = (22,5))
sns.countplot(data = data, x = 'decade', hue = 'body')
plt.title('Body vs Decade')
```

In the period of 1970 - 1980s, sedans were selling more than other body types. In the period of 1980 - 1990s, hatches were getting popular but still sedans were selling more. In 1990 - 2000 period, sales of cars started to grow exponentially with sedans still the most popular followed by hatch and vagon. In 2000 - 2010, sales of sedans skyrocketed to new levels followed by crossovers. After 2010, sales of crossovers overtook the sales of sedans followed by hatches and vans.

This clearly indicates that till 2010, sedans had been extremely popular with people but after 2010, crossovers became equally popular with people.

```python
f,ax = plt.subplots(1,2,figsize = (12,6))
data['body'][data['decade'] == '2000s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
data['body'][data['decade'] == '2010s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1],shadow=True)
ax[0].set_title('Percentage Sales of Car Body Type (2000s)')
ax[1].set_title('Percentage Sales of Car Body Type (2010s)')
```

As seen, in 2000s, 40% cars sold were Sedans while 19.4% cars sold were Crossovers. But in 2010s, Crossovers overtake sales over Sedans at 30.3% vs 29.2%.

How price of cars has increased over the years

```python
mean_price_per_year = data[["year","price"]].groupby("year").mean()
mean_price_per_year
```

```python
mean_price_per_year [['price']].plot(kind='line', figsize=(15,7), color='blue', fontsize=13, linestyle='-.')

plt.xlabel('Year')
plt.ylabel('Avg Price of Cars')
plt.title('Line chart showing the average price of cars by decade')
```

We can see that while the rate of increase in price of cars is quite gradual till 2010, it has increased exponentially after that.

Lets check if count of cars per engType has been over the years

```python
fig,ax = plt.subplots(figsize = (22,7))
sns.countplot(data = data, x = 'decade', hue = 'engType')
plt.title('engType vs Decade')
```

We see that Petrol cars have been preferred by people over Diesel and other fuel cars. But we also see that while in 2000s the number of petrol cars sold is quite high than that of other vehicles but in 2010s the number of diesel cars sold is also quite good. Hence lets see the percentage of various engType cars sold in 2000s and 2010s

```python
f,ax = plt.subplots(1,2,figsize = (12,6))
data['engType'][data['decade'] == '2000s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
data['engType'][data['decade'] == '2010s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1],shadow=True)
ax[0].set_title('Percentage Sales of Engine Types (2000s)')
ax[1].set_title('Percentage Sales of Engine Types (2010s)')
```

So we can see that while Petrol Cars form around 45% of total cars sold in both 2000s and 2010s, the percentage of Diesel Cars and Gas Powered Cars have changed significantly. While in 2000s, Diesel cars formed 28.6% of total sales, in 2010s, it has increased to 39.2% of total sales. On the contrary, while Gas Powered vehicles formed 22.3% of total sales, it gets reduced to just 9% of total sales in 2010s.

We see that a whopping 94.1% of cars are registered in some place whereas only 5.9% cars are not registered

```python
plt.figure(figsize=(5,5))
sns.countplot('drive',data = data,order = data['drive'].value_counts().index)
plt.title("Drive Counts", fontsize=20, fontweight='bold')
plt.xlabel('Drive')
plt.ylabel('Count')
```

```python
data['drive'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True, figsize=(22,6))
plt.xlabel('Drive')
plt.ylabel('')
```

We see that majority of vehicles that have been driven is front drive vehicles at 58%. Next is full drive vehicles at 26.7% and rear wheel drive vehicles at 15.3%.

Let's see if there is a change in drive in the last 3 decades

```python
f,ax = plt.subplots(1,3,figsize = (22,6))
data['drive'][data['decade'] == '1990s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
data['drive'][data['decade'] == '2000s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1],shadow=True)
data['drive'][data['decade'] == '2010s'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[2],shadow=True)
ax[0].set_title('Percentage Sales of Drives (1990s)')
ax[1].set_title('Percentage Sales of Drives (2000s)')
ax[2].set_title('Percentage Sales of Drives (2010s)')
```

We can see how the preference of drives have been changing by decade. While Front Wheel Drives are preferred by Customers at around 57%, Full wheel Drives, which consisted only 8% sales in 1990s, forms 26.1% in 2000s and 36.3% in 2010s. On the contrary, Rear Wheel Drives which used to form 34.2% sales in 1990s, forms 14.3% sales in 2000s and only 8.1% sales in 2010s.

So even though majority of people still prefer front which drives, the preference of full wheel drives has risen significantly over rear wheel drives

Lets check sales of vehicles in terms of mileage

```python
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
data[data['mileage'] != 0]['mileage'].plot(kind='hist', bins = bins, figsize=(15, 7), color='brown', yticks=np.arange(0, 2400, 200), xticks=bins)

plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.title('Histogram showing the distribution of Mileage values')
```

From the above histogram we can see that majority of people prefers those cars which have lower mileages. The maximum sales have happened for vehicles with mileage values in the range of 100-150 and after that from 50-100. Next sales happens for the range of 150-200 and then for 0-50. After the mileage value crosses 200, the sales start to lower gradually.

Hence lets compare the price of vehicle with respect to mileage in the mileage range of 0-200 for last 2 decades (year greater than 2000)

```python
data[(data['mileage'] <= 200) & (data['year'] > 2000) & (data['price'] < 350000)].plot(kind='scatter', x='mileage', y='price', figsize=(15, 7), color='purple', grid=False)

plt.title('Scatter plot showing the variation of Price with Mileage of the Vehicle')
```

We can see that the Price of Car decreases with increase in Mileage of the Car.

Calculating the average price of cars falling in this category:

```python
data[(data['year'] > 2000) & (data['price'] != 0) & (data['mileage'] <= 200)]['price'].median()
```

We can say that if a car has mileage values less than 200 and price tag of around $12,000, customers will be preferred to buy such vehicles
